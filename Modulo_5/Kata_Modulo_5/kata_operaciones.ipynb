Ejercicio 2: convierte cadenas en números y usa valores absolutos

```python
planet1="Mercury"
planet2="Earth"

distancePlanet1=input(f"What is the distance between {planet1} and the sun? ")
distancePlanet2=input(f"What is the distance between {planet2} and the sun? ")
```

```python
distancePlanet1=int(distancePlanet1)
distancePlanet2=int(distancePlanet2)
```

```python
distancebtwplanets=(abs(distancePlanet1-distancePlanet2))
distancebtwplanetsMiles=distancebtwplanets*0.621
```

```python
print(round(distancebtwplanetsMiles))
```
